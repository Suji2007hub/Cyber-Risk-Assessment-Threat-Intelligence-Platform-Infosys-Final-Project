2. Build Threat Scoring Logic

```python
def threat_score(tech):

    #Base score
    score = 5
    name = tech.get("name").lower()
    description = tech.get("description").lower()

    #Name-Based Intelligence
    if "credential" in name:
        score += 3

    if "execution" in name:
        score += 2

    if "privilege" in name:
        score += 3

    if "persistence" in name:
        score += 2

    if "lateral" in name:
        score += 2

    #Description-Based Intelligence
    if "administrator" in description:
        score += 2

    if "remote" in description:
        score += 2

    if "bypass" in description:
        score += 2

    if "stealth" in description:
        score += 1

    return score


```

```python
for tech in techniques[:10]:
    print("Technique:", tech.get("name"))
    print("Score:", threat_score(tech))
```

3. Score All Techniques

```python
scored_techniques = []
for tech in techniques:
  name = tech.get("name", "Unknown")
  score = threat_score(tech)
  scored_techniques.append((name, score))

print("Scoring completed for all techniques")
```

4. Prioritize Threats

```python
scored_techniques.sort(reverse=True)
for item in scored_techniques[:10]:
    name = item[0]
    score = item[1]
    print(f"{name}: {score}")
```

5. Critiical Threat Filter

```python
for name,score in scored_techniques:
    if score >= 8.9:
        print("WArning Critical Threat","Name =",name,",","Score =",score)
```
